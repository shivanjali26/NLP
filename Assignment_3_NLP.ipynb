```python
from collections import defaultdict

def preprocess(text):
    text = text.lower()
    words = text.split()
    return words


def train_ngram_model(words, n):
    model = defaultdict(lambda: defaultdict(int))

    for i in range(len(words) - n + 1):
        context = tuple(words[i:i+n-1])
        next_word = words[i+n-1]
        model[context][next_word] += 1

    return model


def predict_next_word(model, context):
    context = tuple(context)

    if context not in model:
        return "No prediction available"

    next_words = model[context]
    predicted = max(next_words, key=next_words.get)

    return predicted


text = """
Artificial intelligence is transforming the world and artificial intelligence is the future.
Data science and machine learning are important fields in artificial intelligence.
Deep learning models can solve complex problems and deep learning improves accuracy.
"""

words = preprocess(text)

n = 3

model = train_ngram_model(words, n)

contexts = [
    ["artificial" , "intelligence"],
    ["deep", "learning"],
    ["machine", "learning"],
    ["data", "science"]
]

for context in contexts:
    prediction = predict_next_word(model, context)
    print("Context:", context, "-> Predicted:", prediction)
```
