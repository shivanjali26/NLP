```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
input_sentence = "The quick brown fox jumps over the lazy dog. It's a beautiful day!"
```

```python
import re
import nltk
nltk.download('punkt_tab', quiet=True)

processed_text = input_sentence.lower()

processed_text = re.sub(r'[^a-z\s]', '', processed_text)

tokens = word_tokenize(processed_text)

print(f"Original Sentence: {input_sentence}")
print(f"Processed and Tokenized: {tokens}")
```

```python
pos_tags = pos_tag(tokens)

print(f"POS Tags: {pos_tags}")
```

```python
print("\nWord \t\t-> POS Tag")
print("------------------")
for word, tag in pos_tags:
    print(f"{word} \t\t -> {tag}")
```
