```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
```

```python
text_corpus = "\nThis is a sample sentence demonstrating tokenization, stemming, and lemmatization. \nRunning, ran, runs, and quickly are all examples of words we'll process. \nBetter, best, and good are also interesting cases. The cat is running. Cats are running.\n"
```

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text_corpus)
print("Original Text:", text_corpus)
print("\nTokens:", tokens)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in tokens]

print("Original Tokens:", tokens)
print("\nStemmed Tokens:", stemmed_tokens)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word, pos='v') for word in tokens]
lemmatized_tokens_noun = [lemmatizer.lemmatize(word, pos='n') for word in tokens]

print("Original Tokens:", tokens)
print("\nLemmatized Tokens (verbs assumed):\n", lemmatized_tokens)
print("\nLemmatized Tokens (nouns assumed):\n", lemmatized_tokens_noun)

print("\nExamples of Lemmatization with POS tagging:")
print(f"'better' (adj): {lemmatizer.lemmatize('better', pos='a')}")
print(f"'running' (verb): {lemmatizer.lemmatize('running', pos='v')}")
print(f"'cats' (noun): {lemmatizer.lemmatize('cats', pos='n')}")
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]

print("Original Tokens:", tokens)
print("\nFiltered Tokens (after stop words removal):", filtered_tokens)
```
