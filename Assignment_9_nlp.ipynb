```python
import numpy as np
import string
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

text_corpus = """
Machine learning is fascinating and powerful.
Deep learning models can generate text.
Neural networks learn patterns from data.
Text generation is an interesting application of AI.
"""

sequence_length = 3
epochs = 200

# Lowercase
text = text_corpus.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenize
words = text.split()

# Create vocabulary
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
total_words = len(tokenizer.word_index) + 1

# Convert words to integers
encoded = tokenizer.texts_to_sequences([text])[0]

input_sequences = []
output_words = []

for i in range(len(encoded) - sequence_length):
    input_seq = encoded[i:i+sequence_length]
    output_word = encoded[i+sequence_length]

    input_sequences.append(input_seq)
    output_words.append(output_word)

X = np.array(input_sequences)
y = np.array(output_words)

model = Sequential([
    Embedding(input_dim=total_words, output_dim=50, input_length=sequence_length),
    SimpleRNN(100),
    Dense(total_words, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()

model.fit(X, y, epochs=epochs, verbose=1)

def generate_text(seed_text, next_words):
    for _ in range(next_words):
        # Convert seed text to sequence
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=sequence_length, padding='pre')

        # Predict next word
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)[0]

        # Convert index to word
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                seed_text += " " + word
                break

    return seed_text

print("\n===== GENERATED TEXT =====")

seed = "machine learning is"
generated = generate_text(seed, 5)

print("Seed:", seed)
print("Generated:", generated)
```
