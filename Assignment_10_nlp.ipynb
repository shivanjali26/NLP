```python
!pip install gensim
import numpy as np
import string
import nltk

from nltk.corpus import stopwords
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')

Q = "How can I reset my password?"

D = [
    "How to change my password",
    "Steps to reset account password",
    "Internet is not working",
    "Payment issue in my bill",
    "How to contact customer support"
]

stop_words = set(stopwords.words('english'))

def preprocess(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return [word for word in text.split() if word not in stop_words]

# Clean dataset
D_clean = [preprocess(sentence) for sentence in D]
Q_clean = preprocess(Q)

# Train Word2Vec model
model = Word2Vec(sentences=D_clean + [Q_clean], vector_size=100, window=5, min_count=1)

def get_sentence_vector(words):
    vectors = [model.wv[word] for word in words if word in model.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(100)

Q_vector = get_sentence_vector(Q_clean)

sentence_vectors = []
for sentence in D_clean:
    vec = get_sentence_vector(sentence)
    sentence_vectors.append(vec)

similarities = []

for vec in sentence_vectors:
    sim = cosine_similarity([Q_vector], [vec])[0][0]
    similarities.append(sim)


ranked = sorted(zip(D, similarities), key=lambda x: x[1], reverse=True)

print("\n===== SEMANTIC SEARCH RESULT =====")
print("Query:", Q)

print("\nTop Matches:")
for sentence, score in ranked:
    print(f"{sentence}  --> Similarity: {round(score, 3)}")

print("\nMost Relevant Result:", ranked[0][0])

print("\nComment:")
print("Word embeddings capture contextual similarity beyond exact word matching.")
```
