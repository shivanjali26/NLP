```python
!pip install gensim

import nltk
from gensim.models import Word2Vec
import numpy as np

nltk.download('punkt')

corpus = [
    "Artificial intelligence is the future",
    "Machine learning is part of artificial intelligence",
    "Python is used in machine learning",
    "Deep learning is a branch of machine learning",
    "Data science uses python and machine learning",
    "Python programming is easy and powerful",
    "Artificial intelligence improves technology",
    "Machine learning helps in predictions",
    "Deep learning uses neural networks",
    "Data analysis is important in data science"
]

stop_words = {'is', 'in', 'for', 'and', 'the', 'of'}

processed_corpus = []
for sentence in corpus:
    words = sentence.lower().split()
    words = [w for w in words if w not in stop_words]
    processed_corpus.append(words)

model = Word2Vec(
    sentences=processed_corpus,
    vector_size=100,
    window=3,
    min_count=1,
    sg=1,
    epochs=200
)

def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

word1 = "machine"
word2 = "learning"

similarity = cosine_similarity(model.wv[word1], model.wv[word2])

print(f"Similarity between '{word1}' and '{word2}': {similarity}")

print("\nWords similar to 'machine':")
print(model.wv.most_similar("machine"))
```
