```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.chunk import RegexpParser

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt_tab')
```

```python
paragraph = """Machine learning models are widely used in modern applications.
The data scientist builds intelligent systems using large datasets.
Advanced algorithms help in accurate predictions and decision making."""

sentences = nltk.sent_tokenize(paragraph)

grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
chunk_parser = RegexpParser(grammar)
noun_phrases = []
```

```python
for sentence in sentences:
    words = word_tokenize(sentence)
    pos_tags = pos_tag(words)
    tree = chunk_parser.parse(pos_tags)

    for subtree in tree.subtrees():
        if subtree.label() == 'NP':
            np = " ".join(word for word, tag in subtree)
            noun_phrases.append(np)

print("Extracted Noun Phrases:\n")
for np in noun_phrases:
    print("-", np)
```
